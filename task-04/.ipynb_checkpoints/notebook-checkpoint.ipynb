#### Коэффициенты  Пирсона
Рассчитаем коэффициенты  Пирсона  
Для этого воспользуемся формулой:  
$$r_{X, Y} = \frac{\sum\limits_{i=1}^n(X_i - \overline X)(Y_i - \overline Y)}{\sqrt{\sum\limits_{i=1}^n(X_i - \overline X)^2\cdot \sum\limits_{i=1}^n(Y_i - \overline Y)^2}}$$  
  
Назвовем $r_{X, Y}$ коэффициентом корелляции Пирсона или выборочной корелляцией  

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('youtube_4.csv')
df.head()
```

```python
def getR(X, Y):
    xx = df[X].transform(lambda x: x - df[X].mean())
    yy = df[Y].transform(lambda x: x - df[Y].mean())
    numerator = (yy * xx).sum()

    xx = df[X].transform(lambda x: (x - df[X].mean())**2)
    yy = df[Y].transform(lambda x: (x - df[Y].mean())**2)
    denominator = np.sqrt(xx.sum() * yy.sum())

    return numerator/denominator
```

```python
r_coefs = {}
for i in range(2, len(df.columns)):
    for j in range(i, len(df.columns)):
        r = getR(df.columns[i],df.columns[j])
        r_coefs.update({'{}-{}'.format(df.columns[i],df.columns[j]): r})
        print('r', df.columns[i],df.columns[j], r)
```

Таким образом, коэффициенты корреляции Пирсона:  

  
Это и есть кореляционная матрица на основании коэффициентов Пирсона.  
  
Проверим гипотезы о независимости признаков *(об отсутствии корреляции)* с помощью коэффициента корелляции Пирсона.  
  
Пусть  $\rho = \text{Corr}(X_i, Y_i)$ – истинный, то есть теоретический, не выборочный, коэффициент корреляции между  $(X_i,Y_i)$  
  
Сформулируем основную и альтернативную гипотезы:
- $H_0: \rho = 0$ – $X_i$ и $Y_i$ **независимы**  
- $H_A: \rho \neq 0$ – $X_i$ и $Y_i$ **зависимы**  
  
Пусть $X_i \sim N(\mu_X, \sigma_X^2)$ и $Y_i \sim N(\mu_Y, \sigma_Y^2)$ и все С.В. независимы, тогда в качестве статистики выберем  
$$t=\frac{r_{X,Y}\sqrt{n-2}}{\sqrt{1-r_{X,Y}^2}}$$
  
Рассчитаем статистики для всех возможных пар признаков видеороликов:  

```python
def getT(r):
    numerator = r * np.sqrt(100 - 2)
    denominator = np.sqrt(1-r**2)

    return numerator/denominator
```

```python
t_stats = {}
for i in range(2, len(df.columns)):
    for j in range(i + 1, len(df.columns)):
        t = getT(r_coefs['{}-{}'.format(df.columns[i],df.columns[j])])
        t_stats.update({'{}-{}'.format(df.columns[i],df.columns[j]): t})
        print('t', df.columns[i],df.columns[j], t)
```

Таким образом, получаем статистики для каждой пары признаков:  
  
  
Основную гипотезу отвергнем, если $|t| < t_{n-2, \frac\alpha2}$,  
где $\alpha$ – уровень значимости, $t_{n-2, \frac\alpha2}$ – квантиль распределения Стьюдента  
  
Для каждой статистики примем или отвергнем основуню гипотезу на уровнях значимости 5%,1%,0.1%.  
  
Табличные значения:
- $t_{n-2,\frac{\alpha}{2}}=t_{98,\frac{0.05}{2}}=1.98$  
- $t_{n-2,\frac{\alpha}{2}}=t_{98,\frac{0.01}{2}}=2.63$  
- $t_{n-2,\frac{\alpha}{2}}=t_{98,\frac{0.001}{2}}=3.39$  
  
Оценим значимость каждого коэффициента:  
  
  
На диагонали стоят единицы, так как связь характеристики с собой прямая.  
Коэффиценты, не выделенные цветом, незначимы (нет оснований опровергнуть гипотезу об отсутсвии кореляции между этими характеристиками на уровне 0.05)
  
Для характеристик, чьи связи коэффициенты выделены цветом, есть основания опровергнуть гипотезу об отсутствии корелляции, то есть, они зависимы друг от друга.

#### Коэффициенты Спирмена
Рассчитаем коэффициенты ранговой корелляции Спирмена – это коэффициент Пирсона между рангами наблюдений    
Для этого воспользуемся формулой:  
$$r^S_{X,Y}=r_{\text{rank}(X),\text{rank}(Y)}=\frac{\sum\limits_{i=1}^n(\text{rank}(X_i)-\overline{\text{rank}(X)})(\text{rank}(Y_i)-\overline{\text{rank}(Y)})}{\sqrt{\sum_\limits{i=1}^n(\text{rank}(X_i)-\overline{\text{rank}(X)})^2\cdot\sum\limits_{i=1}^n(\text{rank}(Y_i)-\overline{\text{rank}(Y)})^2}}$$  
  
  
Назвовем $r^S_{X, Y}$ коэффициентом ранговой корелляции Спирмена  

```python
for i in range(2, 7):
    df[df.columns[i] + '_rank'] = df[df.columns[i]].rank()
df.head()
```

```python
rs_coefs = {}
for i in range(2, 7):
    for j in range(i, 7):
        r = getR(df.columns[i] + '_rank', df.columns[j] + '_rank')
        rs_coefs.update({'{}-{}'.format(df.columns[i],df.columns[j]): r})
        print('rs', df.columns[i],df.columns[j], r)
```

Таким образом, коэффициенты корреляции Спирмена:  
  
  
Это и есть кореляционная матрица на основании коэффициентов Спирмена  
  
Проверим гипотезы о независимости признаков *(об отсутствии корреляции)* с помощью коэффициента корелляции Спирмена.  
  
Пусть  $\rho = \text{Corr}(X_i, Y_i)$ – истинный, то есть теоретический, не выборочный, коэффициент корреляции между  $(X_i,Y_i)$  
  
Сформулируем основную и альтернативную гипотезы:
- $H_0: \rho = 0$ – $X_i$ и $Y_i$ **независимы**  
- $H_A: \rho \neq 0$ – $X_i$ и $Y_i$ **зависимы**  
  
Пусть в выборке $(X_1,Y_1),..(X_N,Y_N)$ пары $(X_i,Y_i)$ независимы и одинаково распределены.  
Тогда воспользуемся следующей статистикой:  
$$t=\frac{r^S_{X,Y}\sqrt{n-2}}{\sqrt{1-(r^S_{X,Y})^2}} $$

```python
ts_stats = {}
for i in range(2, 7):
    for j in range(i + 1, 7):
        t = getT(rs_coefs['{}-{}'.format(df.columns[i],df.columns[j])])
        ts_stats.update({'{}-{}'.format(df.columns[i],df.columns[j]): t})
        print('ts', df.columns[i],df.columns[j], t)
```

Таким образом, получаем статистики для каждой пары признаков:  
  
  
Основную гипотезу отвергнем, если $|t| < t_{n-2, \frac\alpha2}$,  
где $\alpha$ – уровень значимости, $t_{n-2, \frac\alpha2}$ – квантиль распределения Стьюдента  
  
Для каждой статистики примем или отвергнем основуню гипотезу на уровнях значимости 5%,1%,0.1%.  
  
Табличные значения:
- $t_{n-2,\frac{\alpha}{2}}=t_{98,\frac{0.05}{2}}=1.98$  
- $t_{n-2,\frac{\alpha}{2}}=t_{98,\frac{0.01}{2}}=2.63$  
- $t_{n-2,\frac{\alpha}{2}}=t_{98,\frac{0.001}{2}}=3.39$  
  
Оценим значимость каждого коэффициента:  
  
  
На диагонали стоят единицы, так как связь характеристики с собой прямая.  
Коэффиценты, не выделенные цветом, незначимы (нет оснований опровергнуть гипотезу об отсутсвии кореляции между этими характеристиками на уровне 0.05)
  
Для характеристик, чьи связи коэффициенты выделены цветом, есть основания опровергнуть гипотезу об отсутствии корелляции, то есть, они зависимы друг от друга.
