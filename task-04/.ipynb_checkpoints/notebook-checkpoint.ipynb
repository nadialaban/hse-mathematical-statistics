####   
     
   :  
$$r_{X, Y} = \frac{\sum\limits_{i=1}^n(X_i - \overline X)(Y_i - \overline Y)}{\sqrt{\sum\limits_{i=1}^n(X_i - \overline X)^2\cdot \sum\limits_{i=1}^n(Y_i - \overline Y)^2}}$$  
  
 $r_{X, Y}$        

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('youtube_4.csv')
df.head()
```

```python
def getR(X, Y):
    xx = df[X].transform(lambda x: x - df[X].mean())
    yy = df[Y].transform(lambda x: x - df[Y].mean())
    numerator = (yy * xx).sum()

    xx = df[X].transform(lambda x: (x - df[X].mean())**2)
    yy = df[Y].transform(lambda x: (x - df[Y].mean())**2)
    denominator = np.sqrt(xx.sum() * yy.sum())

    return numerator/denominator
```

```python
r_coefs = {}
for i in range(2, len(df.columns)):
    for j in range(i, len(df.columns)):
        r = getR(df.columns[i],df.columns[j])
        r_coefs.update({'{}-{}'.format(df.columns[i],df.columns[j]): r})
        print('r', df.columns[i],df.columns[j], r)
```

 ,   :  

  
        .  
  
     *(  )*     .  
  
  $\rho = \text{Corr}(X_i, Y_i)$  ,   ,  ,     $(X_i,Y_i)$  
  
    :
- $H_0: \rho = 0$  $X_i$  $Y_i$ ****  
- $H_A: \rho \neq 0$  $X_i$  $Y_i$ ****  
  
 $X_i \sim N(\mu_X, \sigma_X^2)$  $Y_i \sim N(\mu_Y, \sigma_Y^2)$   .. ,       
$$t=\frac{r_{X,Y}\sqrt{n-2}}{\sqrt{1-r_{X,Y}^2}}$$
  
       :  

```python
def getT(r):
    numerator = r * np.sqrt(100 - 2)
    denominator = np.sqrt(1-r**2)

    return numerator/denominator
```

```python
t_stats = {}
for i in range(2, len(df.columns)):
    for j in range(i + 1, len(df.columns)):
        t = getT(r_coefs['{}-{}'.format(df.columns[i],df.columns[j])])
        t_stats.update({'{}-{}'.format(df.columns[i],df.columns[j]): t})
        print('t', df.columns[i],df.columns[j], t)
```

####  
               
   :  
$$r^S_{X,Y}=r_{\text{rank}(X),\text{rank}(Y)}=\frac{\sum\limits_{i=1}^n(\text{rank}(X_i)-\overline{\text{rank}(X)})(\text{rank}(Y_i)-\overline{\text{rank}(Y)})}{\sqrt{\sum_\limits{i=1}^n(\text{rank}(X_i)-\overline{\text{rank}(X)})^2\cdot\sum\limits_{i=1}^n(\text{rank}(Y_i)-\overline{\text{rank}(Y)})^2}}$$  
  
  
 $r^S_{X, Y}$      

```python
for i in range(2, 7):
    df[df.columns[i] + '_rank'] = df[df.columns[i]].rank()
df.head()
```

```python
rs_coefs = {}
for i in range(2, 7):
    for j in range(i, 7):
        r = getR(df.columns[i] + '_rank', df.columns[j] + '_rank')
        rs_coefs.update({'{}-{}'.format(df.columns[i],df.columns[j]): r})
        print('rs', df.columns[i],df.columns[j], r)
```

 ,   :  
  
  
          
  
     *(  )*     .  
  
  $\rho = \text{Corr}(X_i, Y_i)$  ,   ,  ,     $(X_i,Y_i)$  
  
    :
- $H_0: \rho = 0$  $X_i$  $Y_i$ ****  
- $H_A: \rho \neq 0$  $X_i$  $Y_i$ ****  
  
   $(X_1,Y_1),..(X_N,Y_N)$  $(X_i,Y_i)$    .  
   :  
$$t=\frac{r^S_{X,Y}\sqrt{n-2}}{\sqrt{1-(r^S_{X,Y})^2}} $$

```python
ts_stats = {}
for i in range(2, 7):
    for j in range(i + 1, 7):
        t = getT(rs_coefs['{}-{}'.format(df.columns[i],df.columns[j])])
        ts_stats.update({'{}-{}'.format(df.columns[i],df.columns[j]): t})
        print('ts', df.columns[i],df.columns[j], t)
```
