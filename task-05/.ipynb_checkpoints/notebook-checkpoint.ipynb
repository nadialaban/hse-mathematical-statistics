### Оценка регрессии
Для оценки регрессии необходимо воспользоваться методом наименьших квадратов.  
Найдем коэффициенты, при которых сумма квадратов отклонений значений объясняемой величины $Y_i$ от значений линейной функции "длинной" и "короткой" регрессии минимальна.  
  
Будем считать по очереди.  
**Для короткой регрессии:**
  $$\sum^{n}_{i=1}(y_i-(γ_1+γ_2x_i))^2 \longrightarrow min$$  
Берем производные и приравниваем к нулю  
    $$\frac{\partial}{\partial γ_1 }\sum^{n}_{i=1}(y_i-γ_1-γ_2x_i)^2 =-2\sum^{n}_{i=1}(y_i-γ_1-γ_2x_i)=0 $$

   $$\frac{\partial}{\partial γ_2 }\sum^{n}_{i=1}(y_i-γ_1-γ_2x_i)^2 =-2\sum^{n}_{i=1}x_i(y_i-γ_1-γ_2x_i)=0 $$

  
**Для длинной регресии:**
  $$\sum^{n}_{i=1}(y_i-(β_1+β_2x_i+β_3 d_i))^2 \longrightarrow min$$
Найдем все коэффициенты через матрицу:
   $$ \hat β =\begin{pmatrix}
   \hat\beta_{1}\\
  \hat\beta_{2}\\
  \hat \beta_{3}\\
\end{pmatrix} =(X^{T}X)^{-1}X^{T}y\text{ , где  }y =\begin{pmatrix}
   Y_{1}\\
  Y_{2}\\
  ..\\
  Y_{65}\\
\end{pmatrix},X =\begin{pmatrix}
  1& X_{1}  & D_{1} \\
  1& X_{2}  & D_{2}\\
  1& ..     & ..\\
  1& X_{65} & D_{65}\\
\end{pmatrix}$$



  $$\frac{\partial}{\partial β_1 }\sum^{n}_{i=1}(y_i-β_1-β_2x_i-β_3 d_i)^2 =-2\sum^{n}_{i=1}(y_i-β_1-β_2x_i-β_3 d_i)=0$$

  $$\frac{\partial}{\partial β_2 }\sum^{n}_{i=1}(y_i-β_1-β_2x_i-β_3 d_i)^2 =-2\sum^{n}_{i=1}x_i(y_i-β_1-β_2x_i-β_3 d_i)=0$$

   $$\frac{\partial}{\partial β_3 }\sum^{n}_{i=1}(y_i-β_1-β_2x_i-β_3 d_i)^2 =-2\sum^{n}_{i=1}d_i(y_i-β_1-β_2x_i-β_3 d_i)=0$$
  
  
Таким образом, мы получили систему для каждого вида регрессии. Решаем эти системы:

**Для короткой регрессии:**
   $$
   \begin{equation*}
 \begin{cases}
   \hat{γ}_1  = \bar{y}-\hat{ γ_2 }\bar{x}
   \\
 \hat{γ}_2 =\frac{\sum^{n}_{i=1}(x_i-\bar{x})(y_i-\bar{y})}{\sum^{n}_{i=1}(x_i-\bar{x})^2}
 \end{cases}
\end{equation*}
   $$

```python
# Решим систему короткой регрессии
numerator = (tmp_df.x.transform(lambda x: x - tmp_df.x.mean()) * 
      tmp_df.y.transform(lambda x: x - tmp_df.x.mean())).sum()
denominator = (tmp_df.x.transform(lambda x: (x - tmp_df.x.mean())**2)).sum()

gamma_2 = numerator/denominator
gamma_1 = tmp_df.y.mean()-gamma2*tmp_df.x.mean()

print("{:.3f}".format(gamma1))
print("{:.3f}".format(gamma2))
```


**Для длинной регрессии:**
   $$
    \hat β =\begin{pmatrix}
   \hat\beta_{1}\\
  \hat\beta_{2}\\
  \hat \beta_{3}\\
\end{pmatrix} = (X^{T}X)^{-1}X^{T}y
   $$

```python
# Решим систему длинной регрессии
X = tmp_df[['x', 'd']].values.tolist()
for x in X:
    x.insert(0,1)

X = np.matrix(X)
Y = np.matrix(tmp_df.y.transform(lambda y: [y]).tolist())

beta = (X.transpose() * X)**-1*X.transpose()*Y

beta_1 = beta[0,0]
beta_2 = beta[1,0]
beta_3 = beta[2,0]

print("{:.3f}".format(beta_1))
print("{:.3f}".format(beta_2))
print("{:.3f}".format(beta_3))
```

Мы нашли следующие коэффициенты:

 \begin{cases}

 \hat{γ}_2 = 0.648
 \end{cases}
\end{equation*}
   $$

**Для длинной регрессии:**
$$

**Интерпретация коэффициентов**  
Уравнение регрессии не только определяет форму анализируемой связи, но и показывает, в какой степени изменение одного признака сопровождается изменением другого признака.

Коэффициент при $х$, называемый коэффициентом регрессии, показывает, на какую величину в среднем изменяется результативный признак $у$ при изменении факторного признака $х$ на единицу.

### Значения  $R^2$  для обеих моделей
  
Коэффициент детерминации:
$$R^2=\frac{ESS}{TSS}=1-\frac{RSS}{TSS}$$

$$ESS=\frac{1}{n}\sum^{n}_{i=1}(\hat{y}_i-\bar{Y})^2$$

$$TSS=\frac{1}{n}\sum^{n}_{i=1}(y_i-\bar{Y})^2$$


```python
def get_short_estimated(x):
    return gamma_1 + gamma_2 * x

def get_long_estimated(x, d):
    return beta_1 + beta_2 * x + beta_3 * d
```

```python
n = tmp_df.shape[0]

TSS = (1/n*tmp_df.y.transform(lambda y: (y - tmp_df.y.mean())**2)).sum()

y_short_estimated = tmp_df.x.transform(lambda x: get_short_estimated(x))
ESS_short = (1/n * y_short_estimated.transform(lambda x: (x - tmp_df.y.mean())**2)).sum()

y_long_estimated = tmp_df.x.combine(tmp_df.d, lambda x,d: get_long_estimated(x, d))
ESS_long = (1/n * y_long_estimated.transform(lambda x: (x - tmp_df.y.mean())**2)).sum()

R2_short=ESS_short/TSS
R2_long=ESS_long/TSS

print("{:.3f}".format(R2_short))
print("{:.3f}".format(R2_long))
```

**Для короткой регрессии:**  
$ R^2=0.275$

**Для длинной регрессии:**  
$R^2=0.281$

#### ВЫВОД  
Можем видеть, что Коэффициент детерминации для короткой регрессии меньше, чем для длинной, что может говорить о менее точной оценке.

### Прогноз оценки за контрольную работу
По найденным в первом пункте формулам спрогнозируем оценуи за контрольную по обеим моделям для наблюдений 51-65

```python
tmp_df = df.tail(15)
y_short_estimated = tmp_df.x.transform(lambda x: get_short_estimated(x))
y_long_estimated = tmp_df.x.combine(tmp_df.d, lambda x,d: get_long_estimated(x, d))

estimated_df = pd.DataFrame(data={'short':y_short_estimated, 'long':y_long_estimated})
estimated_df
```

### Средняя абсолютная ошибка прогноза  MAE
Высчитаем MAE по следующей формуле  

$$
MAE= \frac{1}{15}∑\limits^{65}_{i=51} |y_i− \hat{y}_{i}|
$$

Вычислим:

```python
n = 15

MAE_short = (1/n * (tmp_df.y - y_short_estimated).abs()).sum()
MAE_long = (1/n * (tmp_df.y - y_long_estimated).abs()).sum()

print("{:.3f}".format(MAE_short)) 
print("{:.3f}".format(MAE_long)) 
```

**Для короткой регрессии:**   
$ MAE=2.084$  
  
**Для длинной регрессии:**  
$ MAE =2.243$  
  
#### ВЫВОД 
Модель короткой регрессии подсказала оценку точнее
