**Решение**  

**1.** Чтобы оценить регрессии воспользуемся методом наименьших квадратов. Те найдем такие коэффициенты, при которых сумма квадратов отклонений значений объясняемой величины $Y_i$ от значений линейной функции "длинной" и "короткой" регрессии минимальна.  
Длинная регрессия:  $\sum^{n}_{i=1}(y_i-(β_1+β_2x_i+β_3 d_i))^2 \xrightarrow[\beta_1 \beta_2 \beta_3]{} min$  
Короткая регрессия: $\sum^{n}_{i=1}(y_i-(γ_1+γ_2x_i))^2 \xrightarrow[\gamma_1 \gamma_2]{} min$  

Нужно взять производные и приравнять их к нулю:  
1) Короткая  
$\frac{\partial}{\partial γ_1 }\sum^{n}_{i=1}(y_i-γ_1-γ_2x_i)^2 =-2\sum^{n}_{i=1}(y_i-γ_1-γ_2x_i)=0 $  
$\frac{\partial}{\partial γ_2 }\sum^{n}_{i=1}(y_i-γ_1-γ_2x_i)^2 =-2\sum^{n}_{i=1}x_i(y_i-γ_1-γ_2x_i)=0 $  
В итоге получили систему:  
$
   \begin{equation*}
 \begin{cases}
   \sum^{n}_{i=1}(y_i-γ_1-γ_2x_i)=0
   \\
 \sum^{n}_{i=1}x_i(y_i-γ_1-γ_2x_i)=0
 \end{cases}
\end{equation*}
   $  
$
   \begin{equation*}
 \begin{cases}
   \hat{γ}_1  = \bar{y}-\hat{ γ_2 }\bar{x}=3.876
   \\
 \hat{γ}_2 =\frac{\sum^{n}_{i=1}(x_i-\bar{x})(y_i-\bar{y})}{\sum^{n}_{i=1}(x_i-\bar{x})^2} = 0.467
 \end{cases}
\end{equation*}
   $  
Итого:  
Оцененное уравнение для короткой регрессии:
$\hat{Y}=3.876+0.467X$  
2) Длинная  
 Найдем все коэффициенты через матрицу:  
   $ \hat β =\begin{pmatrix}
   \hat\beta_{1}\\
  \hat\beta_{2}\\
  \hat \beta_{3}\\
\end{pmatrix} =(X^{T}X)^{-1}X^{T}y$,  где $y =\begin{pmatrix}
   Y_{1}\\
  Y_{2}\\
  ..\\
  Y_{65}\\
\end{pmatrix}$, $X =\begin{pmatrix}
  1& X_{1}  & D_{1} \\
  1& X_{2}  & D_{2}\\
  1& ..     & ..\\
  1& X_{65} & D_{65}\\
\end{pmatrix}$  
$\frac{\partial}{\partial β_1 }\sum^{n}_{i=1}(y_i-β_1-β_2x_i-β_3 d_i)^2 =-2\sum^{n}_{i=1}(y_i-β_1-β_2x_i-β_3 d_i)=0$  
$\frac{\partial}{\partial β_2 }\sum^{n}_{i=1}(y_i-β_1-β_2x_i-β_3 d_i)^2 =-2\sum^{n}_{i=1}x_i(y_i-β_1-β_2x_i-β_3 d_i)=0$  
$\frac{\partial}{\partial β_3 }\sum^{n}_{i=1}(y_i-β_1-β_2x_i-β_3 d_i)^2 =-2\sum^{n}_{i=1}d_i(y_i-β_1-β_2x_i-β_3 d_i)=0$  
Получили систему:  
$
    \hat β =\begin{pmatrix}
   \hat\beta_{1}\\
  \hat\beta_{2}\\
  \hat \beta_{3}\\
\end{pmatrix} =\begin{pmatrix}
  3.896\\
 0.436\\
  1.506\\
\end{pmatrix} 
$  
Итого:  
Оцененное уравнение для длинной регрессии: $\hat{Y}=3.896+0.436X+1.506D$

**Интерпретация коэффициентам**

  Уравнение регрессии не только определяет форму анализируемой связи, но и показывает, в какой степени изменение одного признака сопровождается изменением другого признака.

Коэффициент при $х$, называемый коэффициентом регрессии, показывает, на какую величину в среднем изменяется результативный признак $у$ при изменении факторного признака $х$ на единицу.

```python
import pandas as pd
import math
import scipy
from scipy import stats
import itertools
import numpy as np
```

```python
df = pd.read_csv('C:/Users/User1337/Desktop/hw5/data_6.csv', sep=',')
list_x = [v for v in df['x6'].values.tolist() if not math.isnan(v)]
list_d = [v for v in df['d6'].values.tolist() if not math.isnan(v)]
list_y = [v for v in df['y6'].values.tolist() if not math.isnan(v)]
x_mean = sum(list_x)/len(list_x)
y_mean = sum(list_y)/len(list_y)
d_mean = sum(list_d)/len(list_d)

# короткая #####################################

sum_up=0
sum_down=0
for (x,y) in itertools.zip_longest (list_x,list_y):
    sum_up+=(x-x_mean)*(y-y_mean)
    sum_down+=(x-x_mean)**2

gamma_2=sum_up/sum_down

gamma_1=y_mean-gamma_2*x_mean

print("gamma_1 = {:.3f}".format(gamma_1))
print("gamma_2 = {:.3f}".format(gamma_2))
print('------------')

################################################

# длинная ######################################

y_matr=[]
x_matr=[]
for (x,y,d) in itertools.zip_longest (list_x,list_y,list_d):
    x_matr.append([1,x,d])
    y_matr.append([y])

x_matr=np.matrix(x_matr)
y_matr=np.matrix(y_matr)
x_matr_T=x_matr.transpose()
beta_matr=(x_matr_T*x_matr)**(-1)*x_matr_T*y_matr
beta1=beta_matr[0,0]
beta2=beta_matr[1,0]
beta3=beta_matr[2,0]

print("beta_1 = {:.3f}".format(beta1))
print("beta_2 = {:.3f}".format(beta2))
print("beta_3 = {:.3f}".format(beta3))
print('---------------')
################################################

# 2 пункт ######################################

TSS =0 
ESS_short=0
ESS_long=0
y_short_estimated=[]
y_long_estimated=[]
R2_short=0
R2_long=0
n=len(list_y)
i=0

for (x,y,d) in itertools.zip_longest (list_x,list_y,list_d):
    TSS+=1/n*(y-y_mean)**2

    y_short_estimated.append(gamma_1+gamma_2*x)
    ESS_short+=1/n*(y_short_estimated[i]-y_mean)**2

    y_long_estimated.append(beta1+beta2*x+beta3*d)
    ESS_long+=1/n*(y_long_estimated[i]-y_mean)**2

    i+=1


R2_short=ESS_short/TSS
R2_long=ESS_long/TSS

print("{:.3f}".format(R2_short))
print("{:.3f}".format(R2_long))
print('----------')
################################################

# MAE ##########################################

MAE_short=0
MAE_long=0
for (i) in range(50,65):
    MAE_short+=1/15*abs(list_y[i]-y_short_estimated[i])
    MAE_long+=1/15*abs(list_y[i]-y_long_estimated[i])

print("{:.3f}".format(MAE_short)) 
print("{:.3f}".format(MAE_long)) 
print('-----------')
################################################

#for (i) in range(50,65):
    #print("$$\hat{{y}}-short_{{{}}} = {:.3f} $$".format(i+1,y_short_estimated[i]))

#for (i) in range(50,65):
    #print("$$\hat{{y}}-long_{{{}}} = {:.3f} $$".format(i+1,y_long_estimated[i]))
y_my=beta1+beta2*10+beta3*1
y_my
```
